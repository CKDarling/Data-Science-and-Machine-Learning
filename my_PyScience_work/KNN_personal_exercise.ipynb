# Absenteeism at Work

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
a_at_w = pd.read_csv('Absenteeism_at_work.csv',sep=';')
a_at_w.head()
```

```python
a_at_w = a_at_w.drop('ID',axis=1)
```

```python
a_at_w.info()
```

```python
a_at_w.describe()
```

# We are going to check and see how many options employees are claiming for their absence.

```python
print("Total Count of Reasons for Absenteeism: %s " %(a_at_w['Reason for absence'].value_counts().count()))
```

### KNN requires scaled values, we will alter the datapoints in the following cells.

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(a_at_w.drop('Reason for absence',axis=1))
```

```python
scaled_features = scaler.transform(a_at_w.drop('Reason for absence',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=a_at_w.columns[1:])
```

```python
df_feat.shape
```

```python
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_feat
y = a_at_w['Reason for absence']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.27, random_state=47)
```

# K = 4
##### This is just to make sure the algorithm is performing

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

# Obviously, the low K is not going to have any actual success predicting 28 potential classes. Now we will up the K amount and see how well the model begins to perform. 

# K = 25

```python
knn = KNeighborsClassifier(n_neighbors=25)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

# It appears to only have increased accuracy slightly. Perhaps the data is generally unique per individual enough that the algorithm has a difficutl time sifting through it. 

# Lets try one last time. After all, third time's the charm.  
# K = 28

```python
knn = KNeighborsClassifier(n_neighbors=28)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

# This test made the test worse. I would be willing to say this dataset is not excellent for this form of data.
